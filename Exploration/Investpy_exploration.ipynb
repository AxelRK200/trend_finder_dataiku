```python
import datetime

# using now() to get current time 
current_time = datetime.datetime.now() 

date_jour = str(current_time.day) + "/" + str(current_time.month) + "/" + str(current_time.year)
print(date_jour)
```

```python
df_historical_data = investpy.indices.get_index_historical_data(
    index="Nasdaq 100", 
    country="united states", 
    from_date="15/06/2006", 
    to_date=date_jour
)
```

```python
df_historical_data.head()
```

```python
df_historical_data.tail()
```

```python
df_historical_data.isnull().sum()
```

```python
type(df_historical_data.isnull().sum())
```

```python
# Booleen si colonne ne contient pas de champ vide
test_list = [champ == 0 for champ in df_historical_data.isnull().sum()]
print(test_list)
```

```python
# Si toutes les colonnes sont True, résultat = True
res = all(i for i in test_list)
print(res)
```

```python
df_historical_data.describe()
```

Conformément à describe(), il n'y a pas de min == 0 ou <0.<br />

```python
df_historical_data.describe().min()
```

```python
# Booleen si toutes les colonnes > 0
test_list_zero = [champ > 0 for champ in df_historical_data.describe().min()]
print(test_list_zero)
```

```python
# Si toutes les colonnes sont True, résultat = True
res = all(i for i in test_list_zero)
print(res)
```
